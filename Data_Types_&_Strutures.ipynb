Q.27 . Write a code to create a list of 5 random numbers between 1 and 100, sort it in ascending order, and print the
result.(replaced)

```python
import random
random_numbers = []
for _ in range(5):
    random_numbers.append(random.randint(1, 100))
random_numbers.sort()
print(random_numbers)
```
