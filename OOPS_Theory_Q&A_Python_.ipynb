Q.1 What is Object-Oriented Programming (OOP)

- Object-Oriented Programming (OOP) in Python is a coding paradigm that employs objects and classes to organize code, resulting in modular, reusable, and more manageable structures. This method aids in representing real-world entities by combining data (attributes) and behavior (methods) into one cohesive unit

Q.2 What is a class in OOP

- A class serves as a blueprint or model for producing objects. It is a data type defined by the user that specifies the characteristics (attributes) and actions (methods) common to all objects of that particular type.

Q.3 What is an object in OOP

- An object is a basic, independent code unit that combines activity (methods or functions) and data (attributes or properties).

Q.4 What is the difference between abstraction and encapsulation

- Abstraction hides what an object does, focusing on the interface, while encapsulation bundles data and methods how they work together controlling access with getters/setters for security,

Q.5 What are dunder methods in Python

- Dunder methods, which stand for "double underscore methods," are unique predefined Python functions like __init__ and __add__ that begin and conclude with double underscores. They enable developers to easily integrate Python's built-in functions, operators, and syntax with user-defined objects by customizing their behavior.

Q.6 Explain the concept of inheritance in OOP

- A key idea in object-oriented programming (OOP) is inheritance, which enables a new class to inherit characteristics and behaviors (fields and methods) from an existing class. This method encourages code reuse and maintainability by creating a hierarchical "is-a" relationship between classes (for example, a "Dog" is a "Animal").

Q.7 What is polymorphism in OOP

- In Object-Oriented Programming, polymorphism refers to "many forms," allowing different objects to perform a single action or interface in different ways. This allows for flexibility and code reuse by treating different objects uniformly through a common interface (such as a parent class) while they each carry out their unique implementation.

Q.8 How is encapsulation achieved in Python

- Encapsulation in Python is accomplished by a combination of naming conventions, controlled access using the @property decorator, and access modifiers using single (_) and double (__) underscores.

Q.9 What is a constructor in Python

- When an instance of a class is created in Python, a constructor is a specific method that initializes the attributes (state) of the new object. As soon as the object is created, this function is automatically invoked.

Q.10  What are class and static methods in Python

- In Python, methods that are connected to the class itself instead of an instance of the class (an object) are known as class methods and static methods. They are defined using specific decorators and have different interactions with class or instance data.

Q.11 What is method overloading in Python

- Method overloading is a concept in Python that enables a class to have several methods with the same name but distinct parameters (changing in number, type, or order).

Q.12 What is method overriding in OOP

- In OOP, method overriding allows a subclass to provide its own unique code for a method that is already defined in its superclass, keeping the same name, parameters, and return type to achieve polymorphism (various actions from the same method call) at runtime.

Q.13  What is a property decorator in Python

- Python's built-in @property decorator enables a class function to be accessed and controlled similarly to a class attribute.

Q.14 Why is polymorphism important in OOP

- In object-oriented programming (OOP), polymorphism is crucial because it enables programmers to create adaptable, reusable, and maintainable code. It makes it possible to utilize a single interface, method, or operator with many underlying data types or objects, with the particular behavior decided upon at runtime or compile time.

Q.15  What is an abstract class in Python

- In Python, an abstract class defines a common interface and guarantees consistency among related subclasses, acting as a model for other classes. It frequently has one or more abstract methods that are declared without implementation and cannot be instantiated directly.

Q.16 What are the advantages of OOP

- Better code organization (modularity, encapsulation), reusability through inheritance, flexibility through polymorphism, simpler maintenance and debugging, enhanced security, and scalability are all benefits of object-oriented programming (OOP), which models real-world entities as self-contained "objects". Particularly for large systems, this results in less redundant code, quicker development, and more reliable, manageable software.

Q.17 What is the difference between a class variable and an instance variable

- The primary distinction is that an instance variable is specific to each instance (object), whereas a class variable is shared by all instances of a class.

Q.18 What is multiple inheritance in Python

- A new class (child or derived class) in Python can inherit properties and methods from multiple parent (base) classes in object-oriented programming concept known as multiple inheritance.

Q.19  Explain the purpose of ‘’__str__’ and ‘__repr__’ ‘ methods in Python

- The "dunder" (double underscore) functions in Python that specify how an object is represented as a string are called __str__ and __repr__. The main distinction between them is who they are trying to reach and how much information they offer.

Q.20 What is the significance of the ‘super()’ function in Python

- Python's built-in super() function is essential for comprehending and successfully implementing inheritance. Allowing a child class (subclass) to access and call methods specified in its parent class (superclass) is its main significance.

Q.21 What is the significance of the __del__ method in Python

- When an instance of a class is going to be completed or destroyed, a unique method in Python classes called the __del__ method—also referred to as the destructor—is called.

Q.22 What is the difference between @staticmethod and @classmethod in Python

- The primary difference is that a @staticmethod has no implicit first argument and is unable to access class or instance data, whereas a @classmethod receives the class (cls) as an implicit first parameter and is able to access or modify class state.

Q.23 How does polymorphism work in Python with inheritance

- Method overriding and dynamic method dispatch (runtime polymorphism) are the main ways that polymorphism with inheritance functions in Python. This enables child classes to treat themselves as objects of the same parent type while offering their own unique implementations for methods that are already specified in their parent class.

Q.24 What is method chaining in Python OOP

- In Python OOP, method chaining is a technique that improves code readability and conciseness by executing several methods on the same object in a single, sequential line. Making sure that every method in the chain uses the return self statement to return the object instance itself is crucial to enabling this in a custom class.

Q.25 What is the purpose of the __call__ method in Python

- The purpose of the __call__ method in Python is to allow an instance of a class to be called as a function [2, 3, 4]. When you define this method in a class, you can treat instances of that class just like you would a regular function, using parentheses () to pass arguments and execute code [1, 2, 4].
