Q.1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
  total_even_sum = 0
  for number in numbers:
    if number % 2 == 0:
      total_even_sum += number
  return total_even_sum
```

Q.4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

Q.5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fibonacci(n):
    l = [0,1]
    for i in range(2,n):
        l.append(l[-1]+l[-2])
    return l
if __name__ == "__main__":
    n = 10
    fibo = fibonacci(n)
    print("Fibonacci Series: ",*fibo)
    pass
```

Q.6  Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i
two_powers_gen = powers_of_two(5)
print("Powers of 2 up to exponent 5:")
for power in two_powers_gen:
    print(power)
```

Q.9  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32
celsius_temperatures = [0, 11, 30, 37, 200]
fahrenheit_temperatures_map = map(celsius_to_fahrenheit, celsius_temperatures)
fahrenheit_temperatures = list(fahrenheit_temperatures_map)
print(f"Celsius temperatures: {celsius_temperatures}")
print(f"Fahrenheit temperatures: {fahrenheit_temperatures}")
```

Q.10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    is_not_vowel = lambda char: char not in vowels
    filtered_characters = filter(is_not_vowel, input_string)
    result_string = "".join(filtered_characters)
    return result_string
my_string = "Hi how are you"
string_without_vowels = remove_vowels(my_string)
print(f"Given string: {my_string}")
print(f"String without vowels: {string_without_vowels}")
```
