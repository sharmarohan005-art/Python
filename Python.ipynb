11.	What is the purpose of conditional statements in Python?
- Python conditional statements are used to regulate a program's execution flow according to predetermined criteria. Depending on whether a certain condition evaluates to True or False, they enable a program to make decisions and run various code blocks.

12.	How does the elif statement work?
- The elif statement is the short form of "else if," and it is a conditional statement used in Python to check for multiple conditions sequentially. It functions as part of an if-elif-else structure.
Below is how it work’s:

a. The program first performs the if check, which means it evaluates the condition in the first if statement. If the if condition evaluates to False, it then moves on to the first elif statement and evaluates its condition. If the first elif condition is also False, it then moves on to the next elif statement, if there is one, and evaluates its evaluation. The sequence of elif statements is followed by this procedure. The code block indented beneath an elif statement is executed when the elif condition evaluates to True.
If a True condition is found and its corresponding code block is executed, the program skips any remaining elif or else blocks and exits the entire if-elif-else structure.
The optional else block, if it exists, is run if none of the if or elif conditions evaluate to True.

13.	What is the difference between for and while loops?
-	The key difference is that a for loop is normally used when the number of iterations is known beforehand, whereas a while loop is used when the number of iterations is unknown and depends on a condition being met.

14.	Describe a scenario where a while loop is more suitable than a for loop?
- When considering whether to use a "While Loop" or a "For Loop," it's important to look at the nature of the task at hand. Each type of loop has situations where it is best applied:

Scenario for "While Loop":

- Consider a scenario in which you are reading data from a file all the way to the end. You need a loop that reads the file line by line until there are no more lines to read because you have no idea how many lines it contains. Since the number of iterations required depends on the content of the file, a "While Loop" is better appropriate in this situation.

Steps of the Scenario:
- To track the number of iterations or the current state, start by initializing a variable (e.g., iteration_count = 0).
	Establish a condition flag (condition_met = False, for example) that will change when the task is finished (such as when a file reaches the end).
- Use a "While Loop" that keeps running as long as the condition (condition_met) isn't met.
	Update the condition or the counter after every operation. Increase the iteration count by one, for instance (iteration_count += 1).
- Change the condition flag to break out of the loop (condition_met = True) if at any point the condition for stopping the loop is met, such as reaching the required number of iterations or detecting the end of data.

By following these steps in the scenario described, you can effectively determine when a "While Loop" is more suitable than a "For Loop."
