Q.8	What is a hash table, and how does it relate to dictionaries in Python

- A data structure that effectively stores and retrieves data in key-value pairs is called a hash table. In order to find the matching values in an underlying array or other comparable structure, it employs a hash function to translate keys into numerical indices. Python's built-in “dict” data type is an implementation of a hash table. When you create a dictionary in Python, you are essentially creating a hash table.

They are related in the ways mentioned below:

a.	Key-Value Pairs: Data is stored as key-value pairs in both Python dictionaries and hash tables.

b.	Hashing: A hash value is produced by passing the key through an internal hash function when a key-value pair is added to a Python dictionary. The location of the key-value pair in memory is then decided using this hash value.

c.	Quick Lookups: It is particularly effective to retrieve values from a Python dictionary since keys are hashed to specific locations. Using the hash of the value's key, the dictionary may determine the value's likely position without having to go through every element.

d.	Hashable Keys: A key must be "hashable" in order to be utilized in a Python dictionary and, consequently, in a hash table. This implies that the hash value of the key must remain constant during its lifespan and be unchangeable.

Q.14	How does the “in” keyword work differently for lists and dictionaries

- Python's “in” keyword is used for membership checking, but because lists and dictionaries have different underlying data structures and information storage methods, it behaves differently for both of them. For example Python does a sequential search when we use “in” with a list and until it discovers a match or reaches the end of the list, it iterates through each item in the list starting at the beginning. Whereas Python looks for a key, not a value, when utilizing in with a dictionary. Hash tables are used to implement dictionaries, enabling extremely effective lookups. Python computes a hash value for the key being searched and goes straight to the matching place in the hash table rather than iterating through every element.

Q.17	 Describe the time complexity of accessing elements in a dictionary

- The average time complexity for accessing elements in a dictionary or hash map is typically a constant time. This indicates that when the number of entries in the dictionary increases, the time it takes to obtain a value associated with a particular key does not considerably increase.
The utilization of a hash function is responsible for this efficiency. The hash function transforms a key into an index that links to a particular place in the underlying data structure (often an array) when a key is used to access a value. This eliminates the need to iterate through the full collection and enables direct access to the element.
